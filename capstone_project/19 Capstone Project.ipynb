<h2>Loose Change Calculator - OpenCV</h2>
Included in the files for this course is a photograph of some coins (insert_imagename.jpg). Create a python program that can look at the provided image and estimate how much change there is.<br><br>

Taking the image as input, this program should output another image file with the text "ESTIMATED TOTAL VALUE: {VALUE}, and also circle each coin and denote its estimated value.<br><br>

```python
# Import OpenCV, numpy and matplotlib
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
original_img = cv2.imread('capstone-coins.png', 1)
grey_img = cv2.imread('capstone-coins.png', 0)
print(original_img)
```

```python
grey_img = cv2.GaussianBlur(grey_img, (5,5), 0)
```

```python
cv2.imshow('Coins', grey_img)
cv2.waitKey(0)
```

```python
edges = cv2.Canny(grey_img, 0, 40)
print(edges)
```

```python
cv2.imshow('Edges', edges)
```

```python
cv2.destroyAllWindows()
```

```python
plt.subplot(131), plt.imshow(edges)
plt.title('Edges'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(grey_img)
plt.title('Grey_img'),plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(original_img)
plt.title('Original'),plt.xticks([]), plt.yticks([])
```
